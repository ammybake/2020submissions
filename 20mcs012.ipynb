```
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/mid sem/WCE.jpg')

blur = cv2.GaussianBlur(img,(5,5),0)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```

pic = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)# Plot the image with different kernel sizes
kernels = [5, 11, 17]
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for ind, s in enumerate(kernels):
    pic_b = cv2.blur(img, ksize = (s, s))
    ax = axs[ind]
    ax.imshow(pic_b)
    ax.axis('off')
plt.show()
```

```

pic2 = cv2.GaussianBlur(pic, ksize = (7, 7), sigmaX = 0)   

pic4 = cv2.bilateralFilter(img, 7, sigmaSpace = 75, sigmaColor =75)# Plot the images
images = [pic2,pic4]
fig, axs = plt.subplots(nrows = 1, ncols = 4, figsize = (20, 20))
for ind, p in enumerate(images):
    ax = axs[ind]
    ax.imshow(p)
    ax.axis('off')
plt.show()
```
